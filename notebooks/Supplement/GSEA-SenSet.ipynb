```python
import gseapy as gp
from src.plotter import savefig
```

```python
senset = open("resources/SenSet.txt").read().splitlines()
```

```python
enr = gp.enrich(senset, gene_sets="GO_Biological_Process_2023")
results = enr.results
```

```python
ax = gp.barplot(
    results,
    x="Adjusted P-value",
    top_term=50,
    figsize=(5, 18),
    title="GO Biological Process",
)
ax.grid(axis='x')
savefig('figures/Supplement/GO_SenSet_Complete.pdf')
```
